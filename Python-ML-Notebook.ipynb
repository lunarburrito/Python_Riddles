```python
import math
import pylab
xmin, xmax = -2. * math.pi, 2. * math.pi
n = 1000
x = [0.] * n
y = [0.] * n
dx = (xmax - xmin)/(n-1)
for i in range(n):
    xpt = xmin + i * dx
    x[i] = xpt
    y[i] = math.sin(xpt)**2

pylab.plot(x,y)
pylab.show()
```

```python
#A cardioid is the plane figure described in polar coordinates by r=2a(1+cosθ) for 0≤θ≤2π:
#The equation of a vertical cardioid is r = a ± asinθ.
import matplotlib.pyplot as plt
import numpy as np 
import math
%matplotlib inline 
a=1
ro = []
theta = np.linspace(0,2*math.pi,40)
for i in theta : 
    r = a - a*math.sin(i)
    ro.append(r)
    
rop = np.array(ro)
#print(theta,rop)
fig = plt.figure(figsize=(3,3),dpi = 100)
ax = fig.add_axes([0,0,1,1])
#ax.plot.polar(theta,rop)
plt.polar(theta,rop,color='red')
```
